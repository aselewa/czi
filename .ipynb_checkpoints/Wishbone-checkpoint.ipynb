```python
import wishbone
import pandas as pd
# Plotting and miscellaneous imports
import os
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
log_cpm = pd.read_csv('DGE_Matrices/CZI_MergedAll_LogCPM.csv',sep=',')
```

```python
scdata = wishbone.wb.SCData.from_csv(os.path.expanduser('DGE_Matrices/CZI_MergedAll_LogCPM.csv'), 
                data_type='sc-seq', normalize=False)
```

```python
scdata.data = log_cpm
```

```python
scdata.run_pca()
fig, ax = scdata.plot_pca_variance_explained(ylim=(0, 0.1), n_components=30)
```

```python
NO_CMPNTS = 15
scdata.run_tsne(n_components=NO_CMPNTS, perplexity=100)
```

```python
fig, ax = scdata.plot_tsne()
```

```python
fig = plt.figure(figsize=[5, 4])
scdata.plot_tsne_by_cell_sizes(fig=fig)
```

```python
scdata.run_diffusion_map()
fig, ax = scdata.plot_diffusion_components()
```

```python
# Wishbone class
wb = wishbone.wb.Wishbone(scdata)
wb.run_wishbone(start_cell='GAATAATGTGCC0A', components_list=[1, 2], num_waypoints=150)
```

```python
fig, ax = wb.plot_wishbone_on_tsne()
```

```python
tsne = scdata.tsne
cells = tsne.index.values
d01 = []
d02 = []
d12 = []
d31 = []
d32 = []
d71 = []
d72 = []
d152 = []
for cell in cells:
    if cell[-2:-1] == '0':
        d01.append(cell)
    if cell[-2:-1] == '1':
        d12.append(cell)
    if cell[-3:-2] == '0':
        d02.append(cell)
    if cell[-2:-1] == '3':
        d31.append(cell)
    if cell[-3:-2] == '3':
        d32.append(cell)
    if cell[-2:-1] == '7': 
        d71.append(cell)
    if cell[-3:-2] == '7':
        d72.append(cell)
    if cell[-3:-1] == '15':
        d152.append(cell)
tsne["color"] = ''
tsne.loc[d01,"color"] = 'red'
tsne.loc[d02,"color"] = 'pink'
tsne.loc[d12,'color'] = 'green'
tsne.loc[d31,"color"] = 'blue'
tsne.loc[d32,"color"] = 'cyan'
tsne.loc[d71,"color"] = 'black'
tsne.loc[d72,"color"] = 'grey'
tsne.loc[d152,'color'] = 'orange'
plt.figure(figsize=[8,6])
plt.scatter(tsne.iloc[:,0],tsne.iloc[:,1],color=tsne.loc[:,'color'])
```
